```python
import pandas as pd
```

```python
data = pd.read_csv('./dataset/train.csv')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(data.x,data.y)
```

```python
x = data.x
y = data.y
```

```python
# 序列式模型：keras.Sequential()
model = tf.keras.Sequential()
```

```python
# 添加层 Dense f(x) = ax+b
model.add(tf.keras.layers.Dense(1,input_shape=(1,)))
```

```python
#查看模型 ax+b
model.summary()
```

```python
# 优化算法用 梯度下降算法进行优化
# 损失函数 均方差
model.compile(optimizer='adam',
             loss='mse')
#要做归一化处理，否则结果会很怪
y = tf.clip_by_value(y, 1e-5, 99999.)  # clip to avoid nan in cce
```

```python
history = model.fit(x,y,epochs=20)
```

```python
model.predict(x)
#model.predict(pd.Series([20]))
```

```python
# 梯度下降法：梯度下降法是一种致力于找到函数极值点的算法。
# 所谓学习 便是改进模型参数，以便通过大量训练步骤将损失最小化。
# 有了这个概念，将梯度下降法应用于寻找损失函数的极值便构成了依据输入数据的模型学习。
# 梯度的输出向量是一个有若干偏导数构成的向量，他的每个分量对应函数对输入向量的相应分量的偏导
# 梯度的输出向量表明了在每个位置损失函数增长最快的方向
# 可将它视为表示了在函数的每隔位置向拿个反向移动函数值可以增长
```
